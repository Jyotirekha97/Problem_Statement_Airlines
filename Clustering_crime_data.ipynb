```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.rand(100, 2)
data_norm = (data - np.mean(data, axis=0)) / np.std(data, axis=0)

wcss = []
for i in range(1, 11):
    clustering = AgglomerativeClustering(n_clusters=i)
    clustering.fit(data_norm)
    labels = clustering.labels_
    wcss.append(np.sum((data_norm - np.mean(data_norm[labels], axis=0)) ** 2))

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime['Clust']=x
```

```python
crime
```

```python
crime.iloc[:,1:5].groupby(crime.Clust).mean()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime=pd.read_csv(r'crime_data.csv')
```

```python
crime.info()
```

```python
df=crime.iloc[:,1:5]
```

```python
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
```

```python
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([crime,cl],axis=1)
```
